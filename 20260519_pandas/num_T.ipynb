```python
# 2. 컬럼별 타입·결측치 확인
titanic.isnull().sum() 
```

```python
 # 3. 생존자(survived=1)만 필터링
Survived_df = titanic[titanic['Survived'] == 1]
print(Survived_df.head())
print("생존자 수:", len(Survived_df))
```

```python
# 4. 성별에 따른 생존율 계산

result = titanic.groupby('Sex').agg(
    생존율=('Survived', 'mean'),
    인원=('Survived', 'count')
)

result['생존율'] = result['생존율'] * 100
print(result)
```

```python
# 5. age 결측치를 중앙값으로 채우기

titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median()) #채우기
print("처리 후 Age 결측치:", titanic['Age'].isnull().sum()) #확인
```

```python
# 6. 등급별 평균 나이·요금·생존율

result = titanic.groupby('Pclass').agg(
    평균나이=('Age', 'mean'),
    평균요금=('Fare', 'mean'),
    생존율=('Survived', 'mean')
)

print(result)
```

```python
# 7. 1등급 여성 승객 이름 목록

result = titanic.loc[
    (titanic['Pclass'] == 1) & (titanic['Sex'] == 'female'),
    ['Name', 'Pclass', 'Sex']
]

print(result)
```

```python
# 8. 35세 초과 + 1등급 + 남성 승객 수

result = titanic.loc[
    (titanic['Age'] > 35) & 
    (titanic['Pclass'] == 1) & 
    (titanic['Sex'] == 'male')
]

print(len(result))
```

```python
import numpy as np
np.random.seed(42)
n = 150
shop = pd.DataFrame({
    'customer': np.random.choice(['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price':    np.random.choice([15000,29000,45000,89000,120000], n),
    'rating':   np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice([True, False], n, p=[0.1, 0.9])
})
shop.loc[np.random.choice(shop.index, 8, replace=False), 'rating'] = float('nan')
shop.to_csv('shop_orders.csv', index=False)

df = pd.read_csv('shop_orders.csv')
```

```python
# Q1. 카테고리별 총 매출 내림차순

result = df.groupby('category').agg(
    총매출=('price', 'sum')
).sort_values(by='총매출', ascending=False)

print(result)
```

```python
# Q2. 반품률 가장 높은 카테고리

result = df.groupby('category').agg(
    반품률=('returned', 'mean'),
    주문수=('returned', 'count')
).sort_values(by='반품률', ascending=False)

print(result)
```

```python
# Q3. rating 결측치를 category별 평균으로 채우기

df['rating'] = df['rating'].fillna(
    df.groupby('category')['rating'].transform('mean')
)

print(df['rating'].isnull().sum())
```

```python
# Q4. 고객별 총 주문 금액 TOP 3

result = df.groupby('customer').agg(
    총주문금액=('price', 'sum')
).sort_values(by='총주문금액', ascending=False).head(3)

print(result)
```
